```python
# Question Two

import numpy as np

# i)
sample = np.random.poisson(lam=5, size=1000)
## Our sample has an average of 5 which is also lamda and a size of 1000
```

```python
mean = np.mean(sample)
print(mean)
```

```python
Variance = np.var(sample)
print(Variance)
```

```python
import matplotlib.pyplot as plt
plt.hist(sample, bins=15, color="pink")
plt.title("A histogram To visualize the distribution")
plt.xlabel("Number of customers")
plt.ylabel("Frequecy of the occurance")
plt.show()
```
